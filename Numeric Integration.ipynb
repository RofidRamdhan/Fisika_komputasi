# Code

```python
import numpy as np
```

Fungsi

```python
def f(t):
    g = 9.8
    m = 10
    c = 1
    return np.sqrt(g*m/c)*np.tanh(np.sqrt((g*c/m)*t))

def d2f(t):
    g = 9.8
    m = 10
    c = 1
    return (np.sqrt(g*m/c))*(2*c*g*np.tanh(np.sqrt(c*g*t/m))/m+np.sqrt(c*g*t/m)/t)*((np.tanh(c*g*t/m)**2)-1)/4*t

def d4f(t):
    g = 9.8
    m = 10
    c = 1
    x = np.sqrt(g*m/c)
    y = (np.tanh(g*c/m*t)**2)-1
    fac1 = (c**2*g**2*((np.tanh(np.sqrt(c*g*t/m))**2)-1)*np.tanh(np.sqrt(c*g*t/m)))/m**2
    fac2 = (c**2*g**2*(np.tanh(np.sqrt(c*g*t/m))**3))/2*m**2
    fac3 = (3*c*g*np.sqrt(c*g*t/m)*((np.tanh(np.sqrt(c*g*t/m))**2)-1))/4*m*t
    fac4 = (3*c*g*np.sqrt(c*g*t/m)*np.tanh(np.sqrt(c*g*t/m))**2)/2*m*t
    fac5 = (15*c*g*np.tanh(np.sqrt(c*g*t/m)))/8*m*t
    fac6 = (15*np.sqrt(c*g*t/m))/16*t**2 
    fungsi = (x*y*(fac1+fac2+fac3+fac4+fac5+fac6))/t**2
    return fungsi
```

Metode Integrasi

```python
def trapezoidal(x0,xn,n):
    h = (xn - x0) / n
    integration = f(x0) + f(xn)
    for i in range(1,n):
        k = x0 + i*h
        integration = integration + 2 * f(k)
    integration = integration * h/2
    return integration

def simpson13(x0,xn,n):
    h = (xn - x0) / n
    integration = f(x0) + f(xn)
    for i in range(1,n):
        k = x0 + i*h
        if i%2 == 0:
            integration = integration + 2 * f(k)
        else:
            integration = integration + 4 * f(k)
    integration = integration * h/3
    return integration

def simpson38(x0,xn,n):
    h = (xn - x0) / n
    integration = f(x0) + f(xn)
    for i in range(1,n):
        k = x0 + i*h
        if i%3 == 0:
            integration = integration + 2 * f(k)
        else:
            integration = integration + 3 * f(k)
    integration = integration * 3 * h / 8
    return integration
```

Error

```python
def error_trap(f,x0,xn,n):
    h = (xn-x0)/n
    jumlah_d2f = 0
    for i in range (0,n):
        x = ((x0+i*h)+(x0+(i+1)*h))/2
        jumlah_d2f += f(x)
    error = -(xn-x0)/12*h**2*jumlah_d2f/n
    return error

def error_simp13(f,x0,xn,n):
    h = (xn-x0)/n
    jumlah_d4f = 0
    for i in range (0,round(n/2)):
        x = ((x0+i*h)+(x0+(i+1)*h))/2
        jumlah_d4f += f(x)
    error = -((xn-x0)**5)/(90*n**4)*(jumlah_d4f/n)
    return error

def error_simp38(f,x0,xn,n):
    h = (xn-x0)/n
    jumlah_d4f = 0
    for i in range (0,round(n/2)):
        x = ((x0+i*h)+(x0+(i+1)*h))/2
        jumlah_d4f += f(x)
    error = -(xn-x0)/80*h**4*jumlah_d4f/n
    return error
```
