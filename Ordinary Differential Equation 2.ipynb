# CODE

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def dxdt(x, t, v):
    return v

def d2xdt2( x, t, v, c):
    k = 20
    m = 20
    return (-c*v-k*x)/m
```

```python
def rungekutta4(f, g, lower, upper, step, x0, v0, c):
    t = np.arange(lower,upper+step,step)
    x = [x0]
    v = [v0]
    for i in range(len(t)-1) :
        #orde 1
        k1 = step * f(x[i], t[i], v[i])
        l1 = step * g(x[i], t[i], v[i], c)
        #orde 2
        k2 = step * f(x[i] + 0.5 * step, t[i] + 0.5 * k1, v[i] + 0.5 * l1)
        l2 = step * g(x[i] + 0.5 * step, t[i] + 0.5 * k1, v[i] + 0.5 * l1, c)
        #orde 3
        k3 = step * f(x[i] + 0.5 * step, t[i] + 0.5 * k2, v[i] + 0.5 * l2)
        l3 = step * g(x[i] + 0.5 * step, t[i] + 0.5 * k2, v[i] + 0.5 * l2, c)
        #orde 4
        k4 = step * f(x[i] + step, t[i] + k3, v[i] + l3)
        l4 = step * g(x[i] + step, t[i] + k3, v[i] + l3, c)
        #runge-kutta formula
        x.append(x[i] + (k1+2*k2+2*k3+k4)/6)
        v.append(v[i] + (l1+2*l2+2*l3+l4)/6)
    return x, t, v
```

```python
def tabel(t, x, v):
    tabel = pd.DataFrame({'waktu (t)' : t,'posisi (x)': x, 'kecepatan (v)': v})
    display(tabel)
```

```python
def plot(t, x, v,):
    plt.plot(t, x, 'r', label = 'Posisi (x)')
    plt.plot(t, v,'b', label = 'Kecepatan (v)')
    plt.xlabel('t', fontweight = 'bold') ; plt.ylabel('x / v', fontweight = 'bold')
    plt.legend() ; plt.grid() ; plt.show()
    return
```

```python
x1, t1, v1 = rungekutta4(f = dxdt, g = d2xdt2, lower = 0, upper = 150, step = 0.1, x0 = 1, v0 = 0, c = 5)
x2, t2, v2 = rungekutta4(f = dxdt, g = d2xdt2, lower = 0, upper = 150, step = 0.1, x0 = 1, v0 = 0, c = 40)
x3, t3, v3 = rungekutta4(f = dxdt, g = d2xdt2, lower = 0, upper = 150, step = 0.1, x0 = 1, v0 = 0, c = 200)
```
