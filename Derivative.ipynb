# 2. Mathematical Equation
Turunan Numeric Dengan Pendekatan Limit :
$$f'(a) = \lim_{h\rightarrow 0}\frac{f(a+h)-f(a)}{h}$$
1. Forward
  $$f'(x_i) = \frac{f(x_i+1)-f(x_i)}{h}$$
  $$f''(x_i) = \frac{f(x_i+2)-2f(x_i+1)+f(x_i)}{h^2}$$
  
2. Center
  $$f'(x_i) = \frac{f(x_i+1)-f(x_i-1)}{2h}$$
  $$f''(x_i) = \frac{f(x_i+1)-2f(x_i)+f(x_i-1)}{h^2}$$
  
3. Backward
  $$f'(x_i) = \frac{f(x_i)-f(x_i-1)}{h}$$
  $$f''(x_i) = \frac{f(x_i)-2f(x_i-1)+f(x_i-2)}{h^2}$$

# Code

```python
import numpy as np
import sympy as sp
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

```python
xdata = np.arange(0,0.5,0.1)
fxdata = np.array([0.0000,0.0819,0.1341,0.1646,0.1797], float)

x = np.arange(0,np.pi/2,0.1)
fx = lambda x: np.sin(x)+3
```

```python
def forward_diff1(xdata, f, x, h):
    if type(f) != np.ndarray:
        forward_diff1 = (f(x+h)-f(x))/h
    else:
        forward_diff1 = (f[np.where(xdata == x+h)]-f[np.where(xdata == x)])/h
    return forward_diff1

def forward_diff2(xdata, f, x, h):
    if type(f) != np.ndarray:
        forward_diff2 = (f(x+2*h)-(2*f(x+h))+f(x))/(h**2)
    else:
        forward_diff2 = (f[np.where(xdata == x+2*h)]-(2*f[np.where(xdata == x+h)])+f[np.where(xdata == x)])/(h**2)
    return forward_diff2

def central_diff1(xdata, f, x, h):
    if type(f) != np.ndarray:
        central_diff1 = (f(x+h)-f(x-h))/(2*h)
    else:
        central_diff1 = (f[np.where(xdata == x+h)]-f[np.where(xdata == x-h)])/(2*h)
    return central_diff1

def central_diff2(xdata, f, x, h):
    if type(f) != np.ndarray:
        central_diff2 = (f(x+h)-(2*f(x))+f(x-h))/(h**2)
    else:
        central_diff2 = (f[np.where(xdata == x+h)]-(2*f[np.where(xdata == x)])+f[np.where(xdata == x-h)])/(h**2)
    return central_diff2

def backward_diff1(xdata, f, x, h):
    if type(f) != np.ndarray:
        backward_diff1 = (f(x)-f(x-h))/h
    else:
        backward_diff1 = (f[np.where(xdata == x)]-f[np.where(xdata == x-h)])/h
    return backward_diff1

def backward_diff2(xdata, f, x, h):
    if type(f) != np.ndarray:
        backward_diff2 = (f(x)-(2*f(x-h))+f(x-2*h))/(h**2)
    else:
        backward_diff2 = (f[np.where(xdata == x)]-(2*f[np.where(xdata == x-h)])+f[np.where(xdata == x-2*h)])/(h**2)
    return backward_diff2
```

```python
def interpolasi(x,y):
    n = len(x)
    p = np.poly1d(0.0)
    for i in range(n):
        pt = np.poly1d(y[i])
        for j in range(n):
            if j == i:continue
            fac = x[i]-x[j]
            pt *= np.poly1d([1.0, -x[j]])/fac
        p += pt
    return p

def fungsi(x):
    fungsi = interpolasi(xdata,fxdata)
    return fungsi[0] + fungsi[1]*x + fungsi[2]*x**2 + fungsi[3]*x**3 + fungsi[4]*x**4 
```

```python
def list_diff1(fx,x):
    h = 0.1
    list_forward1, list_central1, list_backward1 = [],[],[]
    for i in range(0,len(x)):
        forward = forward_diff1(xdata == 0,fx, x[i],h)
        central = central_diff1(xdata == 0,fx, x[i],h)
        backward = backward_diff1(xdata == 0,fx, x[i],h)
        list_forward1.append(forward)
        list_central1.append(central)
        list_backward1.append(backward)
    return list_forward1, list_central1, list_backward1

def list_diff2(fx,x):
    h = 0.1
    list_forward2, list_central2, list_backward2 = [],[],[]
    for i in range(0,len(x)):
        forward = forward_diff2(xdata == 0,fx, x[i],h)
        central = central_diff2(xdata == 0,fx, x[i],h)
        backward = backward_diff2(xdata == 0,fx, x[i],h)
        list_forward2.append(forward)
        list_central2.append(central)
        list_backward2.append(backward)
    return list_forward2, list_central2, list_backward2
```

```python
def plot_diff1(list_forward,list_central,list_backward):
    plt.plot(x,list_forward,'-ro', label ='Metode Forward')
    plt.plot(x,list_central,'-go', label ='Metode Central')
    plt.plot(x,list_backward,'-bo', label ='Metode Backward')
    plt.title('Plot Turunan Pertama', fontweight = 'bold')
    plt.xlabel('x', fontweight = 'bold')
    plt.ylabel('y', fontweight = 'bold')
    plt.grid()
    plt.legend()
    plt.show()
    
def plot_diff2(list_forward,list_central, list_backward):
    plt.plot(x,list_forward,'-ro', label ='Metode Forward')
    plt.plot(x,list_central,'-go', label ='Metode Central')
    plt.plot(x,list_backward,'-bo', label ='Metode Backward')
    plt.title('Plot Turunan Kedua', fontweight = 'bold')
    plt.xlabel('x', fontweight = 'bold')
    plt.ylabel('y', fontweight = 'bold')
    plt.grid()
    plt.legend()
    plt.show()
```
