```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
def LeastSquare(x,y):
    sigmax = 0
    sigmay = 0
    for i in range (0,len(x)):
        sigmax += x[i]
        sigmay += y[i]
    xp = sigmax/len(x)
    yp = sigmay/len(y)
    penyebut = 0
    pembilang = 0
    for i in range (0,len(x)):
        penyebut += y[i]*(x[i]-xp)
        pembilang += x[i]*(x[i]-xp)
    konsB = penyebut/pembilang
    konsA = yp-xp*konsB
    return round(konsA,4),round(konsB,4)

def EksponensialFit(x,y):
    n = len(x)
    list = []
    for i in range(0,n):
        ln = np.log(y[i])
        list.append(ln)
    kons_a,konsB = LeastSquare(x,list)
    konsA = np.exp(kons_a)
    return round(konsA,4),round(konsB,4)

def PolynomialFit(x, y, z):
    p = np.array(y,float)
    for i in range(0,len(x)):
        x[i].insert(0,1)
    n = (z*2)-1
    for i in range(2,n):
        for k in range(0,len(x)):
            hasilpangkat = x[k][1]**i
            x[k].insert(len(x)-1,hasilpangkat)
    x_new = np.array(x,float)
    x_trans = x_new.T
    matriksA = np.dot(x_trans,x_new)
    matriksB = np.dot(x_trans,p)
    param = np.linalg.solve(matriksA,matriksB)
    return param

def List_input(f, list_x):
    hasil = []
    n = len(list_x)
    for i in range (0,n):
        nilai_fx = f.subs([(xf,list_x[i])])
        hasil.append(nilai_fx)
    return hasil

def plot(x, y, list_linear, list_exp, list_poly):
    plt.scatter(x,y, color = 'black', label = 'Data Awal')
    plt.plot(x,list_linear, '-g',label ='Linear fit')
    plt.plot(x,list_exp, '-b',label ='Eksponensial fit')
    plt.plot(x,list_poly, '-r',label ='Poly fit')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid()
    plt.show()
    return

def nilai_SSE(list_y,y):
    sigma = 0
    for i in range(0,len(list_y)):
        sigma += (y[i]-list_y[i])**2
    return sigma
    
def multivariat_reg(X,Y):
    for i in range(0,len(X)):
        X[i].insert(0,1)
    X = np.array(X,float)
    x_trans = X.T
    mat1 = np.dot(x_trans,X)
    mat2 = np.dot(x_trans,Y)
    param = np.linalg.solve(mat1,mat2)
    return param

def multivariat_inp(f, X):
    fx = []
    for i in range(0,len(X)):
        fxi = f.subs([(x1,X[i][1]),(x2,X[i][2])])
        fx.append(fxi)
    return fx
```

```python
x = np.array([10,20,30,40,50,60,70,80])
y = np.array([25,70,380,550,610,1220,830,1450],float)
print(x,y)
```

Persamaan Linear Fitting

```python
xf = sp.symbols('x')
konsA_linear,konsB_linear = LeastSquare(x,y)
f_linear = konsA_linear + konsB_linear*xf
display(f_linear)
```

Persamaan Eksponensial Fitting

```python
xf = sp.symbols('x')
konsA_exp,konsB_exp = EksponensialFit(x,y)
f_exp = konsA_exp*sp.exp(konsB_exp*xf)
display(f_exp)
```

Persamaan Polynomial Fitting

```python
x_poly = [[10],[20],[30],[40],[50],[60],[70],[80]]

x0, x1, x2 = sp.symbols('x_0 x_1 x_2')
kons_poly = PolynomialFit(x_poly,y,2)
f_poly = x0+x1*xf+x2*xf**2
f_polynew = f_poly.subs([(x0,float(kons_poly[0])),(x1,float(kons_poly[1])),(x2,float(kons_poly[2]))])
display(f_polynew)
```

Membuat Plot

```python
list_linear = List_input(f_linear,x)
list_exp = List_input(f_exp,x)
list_poly = List_input(f_polynew,x)
```

```python
plot(x, y, list_linear, list_exp, list_poly)
```

```python
SSE_Linear = nilai_SSE(list_linear,y)
print('nilai SSE Linear = ', SSE_Linear)
SSE_Linear = nilai_SSE(list_linear,y)
print('nilai SSE Eksponensial = ', SSE_Linear)
SSE_Poly = nilai_SSE(list_poly,y)
print('Nilai SSE Polynomial = ', SSE_Poly)
```
