# Code

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func( x, y ):
    return x**3-2*x

def y_exact(x): 
    return x**4/4 - x**2 + 1
```

```python
def euler(func, y_exact, x0, y0, start, end, step):
    x = np.arange(start, end+step, step)
    y = np.zeros(len(x))
    y[0] = y_exact(x0)
    list_y_exact = np.zeros(len(x))
    list_y_exact[0] = y0
    error = []
    print('x\ty Euler\ty Exact\tError')
    print('------------------------------')
    for i in range(len(x)-1):
        y[i+1] = y[i] + func(x[i], y[i])*step
        gap = abs(y_exact(x[i+1]) - y[i+1])
        list_y_exact[i+1] = y_exact(x[i+1])
        error.append(gap)
        print('%.4f\t%.4f\t%.4f\t%.4f'% (x[i+1],y[i+1],list_y_exact[i+1], gap))
    return x, y, list_y_exact, gap
```

```python
def rungekutta(func, y_exact, x0, y0, start, end, step):
    x = np.arange(start,end+step,step)
    y = np.zeros(len(x))
    y[0] = y_exact(x0)
    list_y_exact = np.zeros(len(x))
    list_y_exact[0] = y0
    error = []
    print('x\ty RK\ty Exact\tError')
    print('------------------------------')
    for i in range(len(x)-1) :
        k1 = step * func(x[i],y[i])
        k2 = step * func(x[i] + step/2.0, y[i] + k1/2.0)
        y[i+1] = y[i] + k2
        gap = abs(y_exact(x[i+1])-y[i+1])
        list_y_exact[i+1] = y_exact(x[i+1])
        error.append(gap)
        print('%.4f\t%.4f\t%.4f\t%.4f'% (x[i+1],y[i+1],list_y_exact[i+1], gap))
    return x, y, list_y_exact, gap
```
